```python
df_4 = df_4[df_4['호선명'] == '7호선']
df_4['사용월'] = df_4['사용월'].astype(str)
df_4 = df_4[df_4['사용월'].between('202301', '202406')]
```

```python
df_4['지하철역'] = df_4['지하철역'].astype(str)
df_4=df_4.groupby('지하철역').mean(numeric_only=True)

indexes_4 = df_4.index
r = []
for i in range(5):
    total = {"key": "0%d시-0%d시 총 승하차인원" %(i+4, i+5), "value": df_4["0%d시-0%d시 승차인원" %(i+4, i+5)] + df_4["0%d시-0%d시 하차인원" %(i+4, i+5)]}
    r.append(total)

total = {"key": "9시-10시 총 승하차인원", "value": df_4["09시-10시 승차인원"] + df_4["09시-10시 하차인원"]}
r.append(total)

for i in range(6, 19):
    total = {"key": "%d시-%d시 총 승하차인원" %(i+4, i+5), "value": df_4["%d시-%d시 승차인원" %(i+4, i+5)] + df_4["%d시-%d시 하차인원" %(i+4, i+5)]}
    r.append(total)

total = {"key": "00시-01시 총 승하차인원", "value": df_4["00시-01시 승차인원"] + df_4["00시-01시 하차인원"]}
r.append(total)

for i in range(4):
    total = {"key" : "0%d시-0%d시 총 승하차인원" %(i, i+1), "value": df_4["0%d시-0%d시 승차인원" %(i, i+1)] + df_4["0%d시-0%d시 하차인원" %(i, i+1)]}
    r.append(total)

df_5 = pd.DataFrame()
for el in r:
    df_5[el["key"]] =  el["value"]

df_5.index = indexes_4
```

```python
new_indexes = df_5.index
```

```python
df_5.to_csv("data/after/역 시간대별 총 승하차인원.csv", encoding="utf-8-sig")
```

```python
from geopy.geocoders import Nominatim
df_8 = pd.DataFrame()
df_8["역명"] = new_indexes

geolocator = Nominatim(user_agent="hi")
for i, loc in enumerate(df_8["역명"]):
    location = geolocator.geocode(loc)
    if location is not None:
        coord = location[-1]
        df_8.loc[i, '위도'] = coord[0]
        df_8.loc[i, '경도'] = coord[1]
    else:
        df_8.loc[i, '위도'] = None
        df_8.loc[i, '경도'] = None
```

* 아래 쿼리는 지정된 반경 내에서 경찰서, 세무서, 도서관, 소방서, 버스 정류장, 물품보관함, 사회복지시설, 초/중/고/대학교 시설을 모두 검색합니다. 이 쿼리를 Overpass API에 전달하면 해당 시설들의 위치 정보를 JSON 형태로 받을 수 있습니다.

```python
import pandas as pd
import requests

# DataFrame에 열 추가
df_8["버스 정류장 수"] = 0
df_8["경찰서 수"] = 0
df_8["도서관 수"] = 0
df_8["소방서 수"] = 0
df_8["사회복지시설 수"] = 0
df_8["학교 수"] = 0
df_8["구청 수"] = 0
df_8["은행 수"] = 0
df_8["병원 수"] = 0
df_8["일반적 주거 건물의 수"] = 0 # 여기부터는 반경 1km로 잡음
df_8["아파트 수"] = 0

def append_columns(df, amenities, n):
    # 열 이름으로 접근하기 위해 열의 위치 인덱스 찾기
    bus_stop_col = df.columns.get_loc("버스 정류장 수")
    police_col = df.columns.get_loc("경찰서 수")
    library_col = df.columns.get_loc("도서관 수")
    fire_station_col = df.columns.get_loc("소방서 수")
    social_facility_col = df.columns.get_loc("사회복지시설 수")
    school_col = df.columns.get_loc("학교 수")
    town_hall_col = df.columns.get_loc("구청 수")
    bank_col = df.columns.get_loc("은행 수")
    hospital_col = df.columns.get_loc("병원 수")
    residential_col = df.columns.get_loc("일반적 주거 건물의 수")
    apartments_col = df.columns.get_loc("아파트 수")
    
    # amenities가 리스트가 아니라면 리스트로 변환
    if not isinstance(amenities, list):
        amenities = amenities['elements']

    for amenity in amenities:
        tags = amenity.get('tags', {})
        if tags.get('highway') == "bus_stop":
            df.iloc[n, bus_stop_col] += 1
        elif tags.get('amenity') == "police":
            df.iloc[n, police_col] += 1
        elif tags.get('amenity') == "library":
            df.iloc[n, library_col] += 1
        elif tags.get('amenity') == "fire_station":
            df.iloc[n, fire_station_col] += 1
        elif tags.get('amenity') == "social_facility":
            df.iloc[n, social_facility_col] += 1
        elif tags.get('amenity') == "school":
            df.iloc[n, school_col] += 1
        elif tags.get('amenity') == "townhall":
            df.iloc[n, town_hall_col] += 1
        elif tags.get('amenity') == "bank":
            df.iloc[n, bank_col] += 1
        elif tags.get('amenity') == "hospital":
            df.iloc[n, hospital_col] += 1
        elif tags.get('building') == "residential":
            df.iloc[n, residential_col] += 1
        elif tags.get('building') == "apartments":
            df.iloc[n, apartments_col] += 1
        

def search_amenity(df, n):
    # Overpass API URL
    overpass_url = "http://overpass-api.de/api/interpreter"

    # Overpass Query 작성 (500m 반경 내의 시설 검색)
    overpass_query = f"""
    [out:json];
    (
      node["highway"="bus_stop"](around:500, {df.iloc[n]["위도"]}, {df.iloc[n]["경도"]});
      node["amenity"="police"](around:900, {df.iloc[n]["위도"]}, {df.iloc[n]["경도"]});
      node["amenity"="library"](around:900, {df.iloc[n]["위도"]}, {df.iloc[n]["경도"]});
      node["amenity"="fire_station"](around:900, {df.iloc[n]["위도"]}, {df.iloc[n]["경도"]});
      node["amenity"="social_facility"](around:900, {df.iloc[n]["위도"]}, {df.iloc[n]["경도"]});
      node["amenity"="school"](around:900, {df.iloc[n]["위도"]}, {df.iloc[n]["경도"]});
      node["amenity"="townhall"](around:900, {df.iloc[n]["위도"]}, {df.iloc[n]["경도"]});
      node["amenity"="bank"](around:900, {df.iloc[n]["위도"]}, {df.iloc[n]["경도"]});
      node["amenity"="hospital"](around:900, {df.iloc[n]["위도"]}, {df.iloc[n]["경도"]});
      node["building"="apartments"](around:900, {df.iloc[n]["위도"]}, {df.iloc[n]["경도"]});
      node["building"="residential"](around:900, {df.iloc[n]["위도"]}, {df.iloc[n]["경도"]});
    );
    out body;
    """
    
    # API 요청 보내기
    response = requests.get(overpass_url, params={'data': overpass_query})
    
    # JSON 데이터로 변환
    data = response.json()
    
    # amenities가 'elements' 키에 포함되어 있는지 확인
    if 'elements' in data:
        append_columns(df, data['elements'], n)
    else:
        print("No elements found in the API response.")

    print(data)

for m in range(len(df_8)):
    search_amenity(df_8, m)
```

```python
df_8.to_csv("data/after/7호선 상권데이터.csv", encoding='cp949')
```
